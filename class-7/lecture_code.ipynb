```python
#Linear Regression
import seaborn as sns
import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

#Load the dataset
df = sns.load_dataset('tips')
X = df['total_bill'].values.reshape(-1,1)
y = df['tip'].values

#Build relationship between X and y
model = LinearRegression()
model.fit(X,y)

#Predict the values
residuals = model.predict(X) - y
sm.qqplot(residuals, line='s')


```

```python
import numpy as np
import matplotlib.pyplot as plt

#Sample data
x = np.array([1,2,3,4,5])
y = np.array([2,3,5,6,8])
#Fit model
model = LinearRegression()
model.fit(x.reshape(-1,1),y)
#Plot
plt.scatter(x,y, color='blue', label='Data')
plt.plot(x,model.predict(x.reshape(-1,1)), color='red', label='Regression line')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```

```python
# Build the linear regression model to predict house prices
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Define Dataset
square_feet = np.array([1000, 1500, 2000, 1500, 3000])
house_prices = np.array([200000, 250000, 300000, 250000, 400000])

# Build the model / Train the model
model = LinearRegression()
model.fit(square_feet.reshape(-1,1), house_prices)

# Make a prediction
predicted_price = model.predict(np.array([2000]).reshape(-1,1))
print (f'Predicted House price for 2000 sqft is {predicted_price}')

# Plot the data
plt.scatter(square_feet, house_prices, color='blue', label='Actual Data')
plt.plot(square_feet, model.predict(square_feet.reshape(-1,1)), color='red', label='Regression Line')
plt.xlabel('Square Feet')
plt.ylabel('House Price')
plt.title('House Price Prediction')
plt.legend()
plt.show()

```
