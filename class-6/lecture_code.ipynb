```python
# Create a dataframe from a dictionary
import pandas as pd
import numpy as np

employees = {
    'Emp_ID': [101, 102, 103, 104, 105],
    'Name': ['Hamza', 'Ali', 'Ahmed', 'Honda', 'Takada'],
    'Age': [24, 56, 30, np.nan, 29],
    'Salary': [50000, 70000, 60000, 80000, np.nan],
    'Department': ['HR', 'Finance', 'IT', 'HR', 'IT'],
    'Location': ['New York', 'Islamabad', 'Lahore', 'Nagoya', 'Tokyo']
}

df = pd.DataFrame(employees)
print(df)
```

```python
# Handling missing values
df['Age'].fillna(df['Age'].mean(), inplace=True)
df.dropna(subset=['Salary'], inplace=True)

print(df)
```

```python
# Filter employees

it_employees = df[df['Department'] == 'IT']
print('IT Employees:\n', it_employees)

finance_employees = df[df['Department'] == 'Finance']
print('Finance Employees:\n', finance_employees)
```

```python
# Sorting employees by salary
df['Experience'] = np.where(df['Salary'] > 70000, 'Senior', 'Junior')
sorted_df = df.sort_values(by='Salary', ascending=False)
print(sorted_df)
```

```python
# Creating another dataframe for merging
new_hires = {
    'Emp_ID': [106, 107],
    'Name': ['Frank', 'John'],
    'Age': [44, 30],
    'Salary': [65000, 55000],
    'Department': ['HR', 'Finance'],
    'Location': ['France', 'Germany']
}
df_new_hires = pd.DataFrame(new_hires)
print(df_new_hires)
```

```python
# Merging dataframes
df_final = pd.concat([sorted_df, df_new_hires], ignore_index=True)
print(df_final)
```

```python
# Add experience column for the new hires
df_final['Experience'] = np.where(df_final['Salary'] > 70000, 'Senior', 'Junior')
print(df_final)
```

```python
# Matplotlib
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5]
y = [10, 20, 30, 40, 50]

plt.plot(x, y, marker='o', color='r', linestyle='--', label='Line 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line plot')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))

# First subplot
ax[0].plot([1,2,3,4], [10,20,35,60], marker='o', color='r', linestyle='--')
ax[0].set_xlabel('X-axis')
ax[0].set_ylabel('Y-axis')
ax[0].set_title('First subplot')

# Second subplot
ax[1].bar(['A', 'B', 'C', 'D'], [5, 6, 7, 8], color='b')
ax[1].set_xlabel('X-axis')
ax[1].set_ylabel('Y-axis')
ax[1].set_title('Second subplot')
plt.tight_layout()
plt.show()

fig.savefig('plot.png', dpi = 300)
```

```python
x = np.linspace(0, 10, 10)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(8, 6))

plt.plot(x, y1, label='sin(x)', marker='o', color='r', linestyle='-')
plt.plot(x, y2, label='cos(x)', marker='x', color='b', linestyle='--')
plt.xlabel('X-axis', fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Sin and Cos functions', fontsize=14)
plt.legend()
plt.grid(True)
plt.annotate('Peak', xy=(1.57, 1), xytext=(3,1.2), arrowprops=dict(facecolor='black', shrink=0.05))
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

```python
categories = np.random.randint(0, 3, 50)
colors = ['r', 'g', 'b']
plt.scatter(x, y, c=[colors[i] for i in categories])
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.show()
```

```python
markers = ['o', 's', 'D']

for i, marker in enumerate(markers):
    plt.scatter(x[categories == i], y[categories == i], c=colors[i], marker=marker, label=f'Category {i}')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter plot')
plt.legend()
plt.show()
```

```python
z = np.random.rand(50) * 100  # Another variable for scaling
plt.scatter(x, y, s=z, alpha=0.5, c=z, cmap='coolwarm')
plt.colorbar(label="Size Scaling Variable")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Dynamic Scaling and Colormap")
plt.show()
```

```python
# Generate a linear relationship
x = np.linspace(0, 10, 50)
y = 3*x + np.random.randn(50) * 5  # y = 3x + noise
plt.scatter(x, y, color='blue', label="Data")
# Fit a linear trend line
m, b = np.polyfit(x, y, 1)  # First-degree polynomial fit
plt.plot(x, m*x + b, color='red', label="Trend Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Scatter Plot with Regression Trend Line")
plt.show()
```

```python
import seaborn as sns
sns.regplot(x=x, y=y, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Seaborn Regression")
plt.show()
```
